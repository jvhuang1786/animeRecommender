```python
import pandas as pd
import numpy as np
```

```python
credits = pd.read_csv('tmdb_5000_credits.csv')
movies_df = pd.read_csv('tmdb_5000_movies.csv')
```

```python
credits.head()
```

```python
movies_df.head()
```

```python
print(credits.shape)
print(movies_df.shape)
```

```python
#rename movie_id as id 
credits_column_renamed = credits.rename(index = str, columns = {'movie_id': 'id'})
#merge on id
movies_df_merge = movies_df.merge(credits_column_renamed, on = 'id')
movies_df_merge.head()
```

```python
movie_cleaned_df = movies_df_merge.drop(columns = ['homepage', 'title_x', 'title_y', 'status', 'production_countries'])
movie_cleaned_df.head()
```

```python
#no nan values 
movie_cleaned_df.info()
```

```python
#Using weighted average 
#calculate all the components based on the above formula
#W = Weighted average
# R = average for the movies as a number from 0 to 10 (mean) = (Rating)
# v = number of votes for the movie = (votes)
# m = minmum votes required to be listed in the Top 250 (currently 3000)
# C = the mean vote across the whole report (currently 6.9)
# W = (Rv +Cm) / (v+m)
v = movie_cleaned_df['vote_count']
R = movie_cleaned_df['vote_average']
C = movie_cleaned_df['vote_average'].mean()
m = movie_cleaned_df['vote_count'].quantile(0.70)
```

```python
movie_cleaned_df['weighted_average'] = ((R*v) + (C*m))/ (v+m)
```

```python
movie_cleaned_df.head()
```

```python
movie_sorted_ranking = movie_cleaned_df.sort_values('weighted_average', ascending = False)
movie_sorted_ranking[['original_title', 'vote_count', 'vote_average', 
                      'weighted_average', 'popularity']].head(20)
```

```python
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
import seaborn as sns

weighted_average = movie_sorted_ranking.sort_values('weighted_average', ascending = False)
plt.figure(figsize=(12,6))
axis1 = sns.barplot(x = weighted_average['weighted_average'].head(10), 
                   y = weighted_average['original_title'].head(10))
plt.xlim(4,10)
plt.title('Best Movies by average votes', weight = 'bold')
plt.xlabel('Weighted Average Score', weight = 'bold')
plt.ylabel('Movie Title', weight = 'bold')
plt.show()
```

```python
popularity = movie_sorted_ranking.sort_values('popularity', ascending = False)
plt.figure(figsize = (12,6))
ax = sns.barplot(x = popularity['popularity'].head(10), y = popularity['original_title'].head(10),
                data = popularity)

plt.title('Most Popular by Votes', weight = 'bold')
plt.xlabel('Score of Poularity', weight = 'bold')
plt.ylabel('Movie Title', weight = 'bold')
plt.show()
```

```python
popularity.head(10)
```

```python
movie_cleaned_df[['normalized_weight_average', 'normalized_popularity']] = movie_normalized_df
```

```python
movies_df = pd.read_csv('movies.csv', usecols = ['movieId', 'title'], dtype = {'movieId':'int32',
                                                                              'title': 'str'})
ratings_df = pd.read_csv('ratings.csv', usecols = ['userId', 'movieId', 'rating'],
                        dtype = {'userId' : 'int32', 'movieId' : 'int32', 'rating' : 'float32'})
```

```python
movies_df.head()
```

```python
ratings_df.head()
```

```python
df = pd.merge(ratings_df, movies_df, on = 'movieId')
df.head()
```

```python
combine_movie_rating = df.dropna(axis = 0,subset = ['title'])
movie_ratingCount = (combine_movie_rating.groupby(by = ['title'])['rating'].count().reset_index().
                    rename(columns = {'rating': 'totalRatingCount'})
                    [['title', 'totalRatingCount']]
                    )
movie_ratingCount.head()
```

```python
rating_with_totalRatingCount = combine_movie_rating.merge(movie_ratingCount, left_on = 'title', 
                                                     right_on = 'title',
                                                    how = 'left')
rating_with_totalRatingCount.head()
```

```python
pd.set_option('display.float_format', lambda x: '%.3f' % x)
print(movie_ratingCount['totalRatingCount'].describe())
```

```python
popularity_threshold = 50
rating_popularity_movie = rating_with_totalRatingCount.query('totalRatingCount >= @popularity_threshold')
rating_popularity_movie.head()
```

```python
rating_popularity_movie.shape
```

```python
## first lets create a pivot table

movie_features_df = rating_popularity_movie.pivot_table(index = 'title', columns = 'userId',
                                                    values = 'rating').fillna(0)
movie_features_df.head()
```

```python
from scipy.sparse import csr_matrix

movie_features_df_matrix = csr_matrix(movie_features_df.values)

from sklearn.neighbors import NearestNeighbors

model_knn = NearestNeighbors(metric = 'cosine', algorithm = 'brute')
model_knn.fit(movie_features_df_matrix)
```

```python
#cosine similarity
# sim(A,B) = cos(theta) = (A*B)/(||A||||B||)
```

```python
movie_features_df.shape
```

```python
query_index = np.random.choice(movie_features_df.shape[0])
print(query_index)
distances, indices = model_knn.kneighbors(movie_features_df.iloc[query_index,:].values.reshape(1,-1),
                                         n_neighbors = 6)
```

```python
movie_features_df.head()
```

```python
#flatten the distance
for i in range(0, len(distances.flatten())):
    if i == 0:
        print('Recommendations for {0}:\n'.format(movie_features_df.index[query_index]))
    else:
        #number, moviename, distance
        print('{0}: {1}, with distance of {2}:'.format(i, movie_features_df.index[indices.flatten()[i]],
                                                      distances.flatten()[i]))
```
