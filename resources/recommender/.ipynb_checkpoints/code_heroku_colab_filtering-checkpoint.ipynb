```python
import pandas as pd
from scipy import sparse
from sklearn.metrics.pairwise import cosine_similarity
```

```python
ratings = pd.read_csv('toy_dataset.csv', index_col = 0)
ratings = ratings.fillna(0)
ratings
```

```python
def standardize(row):
    new_row = (row - row.mean()) / (row.max() - row.min())
    return new_row

ratings_std = ratings.apply(standardize)
```

```python
ratings_std
```

```python
#we are taking a transpoe since we want similarity between items which need to be in rows
item_similarity = cosine_similarity(ratings_std.T)
print(item_similarity)
```

```python
item_similarity_df =  pd.DataFrame(item_similarity, index = ratings.columns, columns = ratings.columns)
```

```python
item_similarity_df
```

```python
#lets make a recommendations

def get_similar_movies(movie_name, user_rating):
    similar_score = item_similarity_df[movie_name] * (user_rating -2.5)
    similar_score = similar_score.sort_values(ascending = False)
    return similar_score

print(get_similar_movies('romantic3', 1))
```

```python
action_lover = [('action1', 5), ('romantic2', 1), ('romantic3',1)]

similar_movies = pd.DataFrame()

for movie,rating in action_lover:
    similar_movies = similar_movies.append(get_similar_movies(movie,rating), ignore_index = True)
    
similar_movies.head()
similar_movies.sum().sort_values(ascending = False)
```

```python
ratings = pd.read_csv('ratings.csv')
movies = pd.read_csv('movies.csv')
ratings = pd.merge(movies, ratings).drop(['genres', 'timestamp'],axis =1)
ratings.head()
```

```python
user_ratings = ratings.pivot_table(index = ['userId'], columns = ['title'], values = 'rating')
```

```python
user_ratings.head()
```

```python
## Let's remove Movies which have less than 10 users who rated it. and fill remaining NaN with 0
user_ratings = user_ratings.dropna(thresh = 10, axis =1).fillna(0)
```

```python
user_ratings.head()
```

```python
#let's build our Similarity Matrix
item_similarity_df = user_ratings.corr(method = 'pearson')
item_similarity_df.head()
```

```python
action_lover = [('2 Fast 2 Furious (Fast and the Furious 2, The) (2003)', 5),
               ('12 Years a Slave (2013)', 4), ('2012 (2009)', 3),
               ('(500) Days of Summer (2009)', 2)]
similar_movies = pd.DataFrame()

for movie, rating in action_lover:
    similar_movies = similar_movies.append(get_similar_movies(movie, rating), ignore_index = True)
    
similar_movies.head()
similar_movies.sum().sort_values(ascending = False)
```
